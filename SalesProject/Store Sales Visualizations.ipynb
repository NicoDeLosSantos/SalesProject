```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sales_df = pd.read_csv("C:/Users/nicod/Desktop/o/PROGRAMACION/Base de datos/SalesProject/Stores.csv")

sales_df
```

```python
sales_df.set_index("Store ID ", inplace=True)
sales_df
```

```python
## sorting by Store_Sales desc
sales_df.sort_values('Store_Sales', ascending=False)
```

```python
columns=['Store_Area','Items_Available','Daily_Customer_Count','Store_Sales']
```

```python
sales_df[columns].describe()
```

```python
top_10_sales = sales_df.sort_values('Store_Sales', ascending=False).head(10)
```

```python
for f in columns:
    fig = plt.figure(figsize = (10,5))
    sns.barplot(x='Store ID ', y=f, data=top_10_sales)
    plt.grid()
    plt.title(f)
    plt.show()
```
